### ¿Qué? ¿Por qué? y ¿Cómo?

#### ¿Qué?

En el anterior gráfico de redes lo que se busco fue observar la relación entre los adjetivos y sustantivos comunes en las novela David Copperfield de Chares Dickens, esto se usa por medio de Node-Links, este diagrama es el que nos permite de la manera más simple visualmente hablando, analizar las relaciones que tenemos entre los los adjetivos y sustantivos por medio de los enlaces. 

#### ¿Por qué?

Este tipo de diagramas son bastante funcionales para cuando lo que queremos es ver las relaciones, en este caso, podemos analizar la foma en que el autor relaciona los adjetivos y sustantivos, nos puede dar una idea general del texto, del cómo se expresaba en su literatura, por eso la **topologia** es interesante de ver de este modo. Sirve para observar las conexiones más fuertes o directas y las más indirectas.

#### ¿Cómo?

Por medio del grafico de redes.
